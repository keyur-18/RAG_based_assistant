```python
import os 
import math
import json
```

```python
n = 5
for file_name in os.listdir("jsons"):
    if file_name.endswith(".json"):
        with open(f"jsons/{file_name}","r",encoding="utf-8")as f:
            data = json.load(f)
            new_chunk = []
            num_chunks = len(data["chunks"])
            num_grps = math.ceil(num_chunks/n)
            for i in range(num_grps):
                start_idx = i*n
                end_idx = min((i+1)*n,num_chunks)

                chunk_grp = data["chunks"][start_idx:end_idx]
                new_chunk.append({
                    'number':data['chunks'][0]['number'],
                    "title" : data['chunks'][0]['title'],
                    "Start" : chunk_grp[0]['Start'],
                    "end" : chunk_grp[-1]['end'],
                    "text" : " ".join(d['text'] for d in chunk_grp)
                })
            os.makedirs("merge_json",exist_ok=True)
            with open(f"merge_json/{file_name}","w",encoding="utf-8")as json_file:
                json.dump({"chunks":new_chunk,"text": data['text']},json_file,indent=4)
```
