```python
import whisper
import ffmpeg
```

```python
model = whisper.load_model("large-v2")
```

```python
import os
audios = os.listdir("/kaggle/input/random-audios")
```

```python
import json
```

```python
for audio in audios:
    number = audio.split(".")[0].split("_")[1]
    title = audio.split("_")[0]
    result = model.transcribe(audio=f"/kaggle/input/random-audios/{audio}",
                             language="hi",
                             task="translate",
                             word_timestamps=False)
    chunks=[]
    for segment in result["segments"]:
        chunks.append({"number":number,"title":title,"Start":segment["start"],"end":segment["end"],"text":segment["text"]})
    chunks_with_metadeta = {"chunks":chunks,"text":result["text"]}
    with open(f"/kaggle/working/{number}_{title}.json","w") as f:
        json.dump(chunks_with_metadeta,f)

    
```
